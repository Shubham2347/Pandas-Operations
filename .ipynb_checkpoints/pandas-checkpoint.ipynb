# Conditional Filtering

#### == ,!=, >, <, >,=, <=

##### single conditions

```python
df["total_bill"]>20
```

```python
df["size"]==3
```

```python
bool_val=df["total_bill"]>45
df[bool_val]
```

```python
df[df["size"]>3]
```

#### Multiple conditions
#### And &--->  both condition need to be true
#### Or |  ----->  either  condition is true

```python
df[(df["sex"]=="Male") & (df["total_bill"]>40) & (df["smoker"]=="Yes")]
```

```python
df[(df["day"]=="Sun") | (df["day"]=="Sat")]
```

```python
                                          #Or
```

```python
options=["Sat","Sun"]
```

```python
df["day"].isin(options)
```

```python
df[df["day"].isin(options)]
```

# Groupby Operation

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("mpg.csv")
```

```python
df
```

```python
df["model_year"].unique()
```

```python
df["model_year"].value_counts()
```

```python
df.groupby("model_year").mean()["mpg"]
```

```python
df.groupby(["model_year","cylinders"]).mean()
```

```python
df.groupby(["model_year","cylinders"]).mean()["mpg"]
```

```python
df.groupby("model_year").describe()
```

```python
df.groupby("model_year").describe().transpose()
```

```python
df[df["cylinders"].isin([6,8])].groupby(["model_year","cylinders"]).mean()
```
